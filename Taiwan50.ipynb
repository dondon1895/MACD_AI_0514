```
import csv
with open('Taiwan50.csv', newline='') as csvfile:

  # 讀取 CSV 檔內容，將每一列轉成一個 dictionary
  rows = csv.DictReader(csvfile)

  # 以迴圈輸出指定欄位
  for row in rows:
    if (row['SMA10'] > row['SMA23'] and row['K(9,3)'] > row['D(9,3)'] and row['DIF6-12'] > row['MACD3']):
      print(row['時間'], row['SMA10'], row['SMA23'],row['K(9,3)'], row['D(9,3)'], row['DIF6-12'] ,row['MACD3'])
```
